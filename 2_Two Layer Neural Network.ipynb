<h1 style="color:navy">II. Two-Layer Neural Network</h1>

İki katmanlı yapay sinir ağı, diğer çok katmanlı sinir ağları gibi temel olarak üç bölümden oluşur. **Input-Layer** (Giriş katmanı), **Hidden-Layer** (Gizli katman) ve **Output-Layer** (Çıkış katmanı). Katman sayısı belirlenirken, input harici tüm katmanlar sayılır. İki katmanlı yapay sinir ağı (YSA), Lojistik Regresyonda kullandığımız en basit sinir ağına, bir katman daha (hidden layer) eklenerek oluşturulmuştur. Hidden ve Output katmanlarındaki nöron sayıları, ağdan ağa farklılık gösterebilir. İki katmanlı YSA, lojistik regresyonun çözmekte yetersiz kaldığı problemlere daha iyi çözümler getirebilir. Katman sayısı arttıkça genelde ağın problem çözme yetisi de artmaktadır. Fakat bu her zaman doğru değildir. Bir probleme daha az katmanla, daha iyi çözümler getiren sinir ağları da bulunabilir.

<img src="https://raw.githubusercontent.com/zkcplk/Deep-Learning/main/images/7_two_layer_neural_network.png" width="800px">

Lojistik Regresyonda kullandığımız tek aktivasyon fonksiyonu sigmoid idi. Burada ise gizli katmanda **hiperbolik tanjant (tanh)** fonksiyonu kullanılır. Bunun sebebi, neredeyse her zaman sigmoid fonksiyonundan daha iyi çalışmasıdır. Sigmoid fonksiyonu 0-1 aralığında değer alırken, Tanh fonksiyonu -1 ile +1 aralığında değer alır. Bazen bir derin öğrenme algoritmasını eğitmek istediğinizde, verilerinizi merkeze almak isteyebilirsiniz. Verileri merkeze almanın etkisi, verilerinizin ortalamasının 0.05 gibi veya 0'a yakın bir değer almasıdır. Bu da bir sonraki katman için öğrenmeyi kolaylaştırır. Fakat çıkış katmanı için hala sigmoid kullanmaya devam edebiliriz. Özellikle çıktının 0 veya 1 olmasını istediğimiz ikili sınıflandırma problemimiz varsa, mutlaka çıkış katmanında sigmoid fonksiyonunu kullanmamız gerekir. Sigmoid ve Tanh dışında daha derin sinir ağlarında, hem hesaplama hızını arttırmak hem de başka faydaları sebebiyle ReLU ve Leaky ReLU isimli aktivasyon fonksiyonları da kullanılır. Şimdilik bunlara değinmiyoruz.

<img src="https://raw.githubusercontent.com/zkcplk/Deep-Learning/main/images/8_sigmoid_tanh_functions.jpg" width="800px">

Katmanları sayısal olarak ifade ederken köşeli parantez kullanırız. **[1]** ifadesi birinci katmanı belirtir. Yukarıdaki örnekte gizli katman, aynı zamanda birinci katmandır. Birinci katmana ait $a$  değerleri, katmanda kullanılan **nöron sayısına göre değişkenlik göstermekle birlikte** şu şekilde sıralanır: 

$$a_1^{[1]} , a_2^{[1]} ,  a_3^{[1]} ...$$



Bu örnekte **gizli katmandaki her bir $a$ değeri** aslında bir tanh fonksiyonunun çıktısıdır: $a_{1,2,3,4}^{[1]}$ = $g$($z_{1,2,3,4}^{[1]}$) = $tanh$($z_{1,2,3,4}^{[1]}$)

Çıktı katmanında ise: $a_{1}^{[2]}$ = $g$($z_{1}^{[2]}$) = $sigmoid$($z_{1}^{[2]}$)

$a_1^{[1]}$ $a_2^{[1]}$ $a_3^{[1]}$ $a_4^{[1]}$ ... bir vektör matrisi oluşturur ve $a^{[1]}$ şeklinde ifade edilir. $a^{[1]}$ gibi $z^{[1]}$ ve $b^{[1]}$ matrisleri de aynı mantıkla yazılabilir.


$$a^{[1]} = 
\begin{bmatrix}
    a_1^{[1]}\\
    a_2^{[1]}\\
    \vdots\\
    a_4^{[1]}
\end{bmatrix} \text{    } z^{[1]} = 
\begin{bmatrix}
    z_1^{[1]}\\
    z_2^{[1]}\\
    \vdots\\
    z_4^{[1]}
\end{bmatrix} \text{    } b^{[1]} = 
\begin{bmatrix}
    b_1^{[1]}\\
    b_2^{[1]}\\
    \vdots\\
    b_4^{[1]}
\end{bmatrix}$$


Bir katmandaki her bir nöronda, o katmana giriş yapan x değerleri (değişkenleri) kadar $w$ değerleri bulunur. Örneğin, üstte verilen YSA incelenirse, 1. katmandaki her bir nöronda, o nörona giren $x_1$ ve $x_2$'den dolayı, $w_1$ ve $w_2$ değerleri vardır ve 1.  katmanda toplam 4 nöron olduğu için, $W^{[1]}$ matrisi (4x2)'lik bir matris olur.


$$W^{[1]} = 
\begin{bmatrix}
    w_1^{[1][1]} w_1^{[1][2]}\\
    w_2^{[1][1]} w_2^{[1][2]}\\
    \vdots\\
    w_4^{[1][1]} w_4^{[1][2]}
\end{bmatrix} \text{    }$$



Giriş katmanındaki $x_1$ $x_2$ $x_3$ ... değerleri ile bir vektör matrisi oluşturulursa, o da $a^{[0]}$ olarak ifade edilebilir. Aşağıda, her katmanda hesaplanan değerlerin matris boyutları verilmiştir.


$$a^{[1]} : (4,1)$$

$$z^{[1]} : (4,1)$$
<br>$$b^{[1]} : (4,1)$$

$$W^{[1]} : (4,2)$$

$$a^{[2]} : (1,1)$$

$$z^{[2]} : (1,1)$$

$$b^{[2]} : (1,1)$$

$$W^{[2]} : (1,4)$$



Aşağıdaki resimde daha önce işlediğimiz Lojistik Regresyon ve şuan incelediğimiz İki katmanlı YSA'da gerçekleştirilen forward ve backward propagation işlemleri, karşılaştırmalı olarak gösterilmiştir.

<img src="https://raw.githubusercontent.com/zkcplk/Deep-Learning/main/images/9_logistic_regression_two_layer_neural_network.png">

Daha önce lojistik regresyonda backward propagation yaparken, kullandığımız türevleme işlemleri, iki katmanlı ve daha sonra işleyeceğimiz çok katmanlı sinir ağlarında da geçerlidir. Türev almaktaki amaç, yeni $w$ ve $b$ değerleri üretmektir. Burada iki katman olduğu için her katmanda ayrı ayrı $w$ ve $b$ değerleri üretilecektir. Dolayısıyla kodumuzda, iki adet **dw** ve iki adet **db** türev değişkeni bulunacaktır.

Yukarıdaki resimde her bir katman bir sonraki katmana bilgi taşımaktadır. Örneğin, birinci katmanda elde edilen $a^{[1]}$ değeri, ikinci katmanda $z^{[2]}$ denkleminde kullanılmıştır. Eğer örneğimizde bir katman daha olsaydı, ikinci katmandan üretilen $a^{[2]}$ değeri, bir sonraki katmanda $z^{[3]}$ denkleminde kullanılacaktı. 

Yazacağımız kodda kullanacağımız formüller aşağıdaki gibidir. Önceki derste ifadelerle ilgili ayrıntılı açıklama yapıldığı için burada sadece formülleri vererek geçeceğiz. Örneğimizdeki YSA'da, veri setindeki her bir veri (i) için **z**, **a** ve **Cost** değeri aşağıdaki gibi hesaplanır:



$$z^{[1] (i)} =  W^{[1]} x^{(i)} + b^{[1]}\tag{1}$$ 
    
$$a^{[1] (i)} = tanh(z^{[1] (i)})\tag{2}$$
    
$$z^{[2] (i)} = W^{[2]} a^{[1] (i)} + b^{[2]}\tag{3}$$
    
$$\hat{y}^{(i)} = a^{[2] (i)} = sigmoid(z^{ [2] (i)})\tag{4}$$
    
$$ \mathcal{L}(a^{[2] (i)}, y^{(i)}) =  - y^{(i)}  \log(a^{[2] (i)}) + (1-y^{(i)} )  \log(1-a^{[2] (i)}) \tag{5}$$

$$J = - \frac{1}{m} \sum\limits_{i = 0}^{m} \mathcal{L}(a^{[2] (i)}, y^{(i)}) \tag{6}$$



Şimdi de backward propagation'da işimize yarayacak diğer ifadeleri yazalım. Bu ifadeler, yukarıdaki şekilde anlatılan türev işlemlerinden sonra elde edilmiştir. Matematiksel ayrıntı konumuz dışında olduğu ve uzun süreceği için direkt formüller üzerinden ilerleyeceğiz. Kod içerisinde hangi formülün nerede kullanıldığı belirtilecektir.
    


$$ dZ^{[2]} = A^{[2]} - Y \tag{7}$$

$$ dW^{[2]} = \frac{1}{m} dZ^{[2]} A^{[1]^{T}} \tag{8}$$
    
$$ db^{[2]} = \frac{1}{m} np.sum(dZ^{[2]},axis=1,keepdims=True) \tag{9}$$
    
$$ dZ^{[1]} = W^{[2]^{T}} dZ^{[2]} * g^{[1]^{`}}(Z^{[1]}) \tag{10}$$

$$ dW^{[1]} = \frac{1}{m} dZ^{[1]} X^{T} \tag{11}$$

$$ db^{[1]} = \frac{1}{m} np.sum(dZ^{[1]},axis=1,keepdims=True) \tag{12}$$



2 katmanlı YSA'nın teorisine değindikten sonra, kodlamalarımızda kullanacağımız fonksiyonları oluşturalım.

```python
# numpy & pandas.
import numpy as np
import pandas as pd

#====================
# sigmoid fonksiyonu.
def sigmoid(z):
    return 1/(1+np.exp(-z))

#====================
# başlangıç parametrelerini belirlemek.
# iki katmanlı bir YSA olduğu için hidden layer, 1 katmandan oluşuyor.
# hidden değeri gizli katmandaki nöron sayısını ifade ediyor.
def initialize_parameters(X, Y, hidden):
    # input layer
    n_x = X.shape[0]
    
    # hidden layer
    n_h = hidden
    
    # output layer
    n_y = Y.shape[0]
    
    # başka kişilerle / başka zamanda aynı sonuçları bulabilmek için aynı random ayarını yapmalıyız.
    np.random.seed(2) 
    
    # lojistik regresyonda başlangıç w ve b değerlerini 0 verebiliyorduk.
    # fakat çok katmanlı YSA için bu geçerli değildir, sıfırdan farklı bir değer vermek gerekir.
    # nöron sayısı arttıkça bulunacak random sayıların değeri de büyüyebilecektir.
    # o yüzden 0.01 ile çarparak, çıkacak olan değeri iyice küçültmüş oluyoruz.
    W1 = np.random.randn(n_h, n_x) * 0.01
    b1 = np.zeros(shape=(n_h, 1))
    W2 = np.random.randn(n_y, n_h) * 0.01
    b2 = np.zeros(shape=(n_y, 1))
    
    # sonuç
    parameters = {"W1": W1,
                  "b1": b1,
                  "W2": W2,
                  "b2": b2}
    
    return parameters

#====================
# yine lojistik regresyondan farklı olarak 
# forward ve backward propagation işlemlerini ayrı ayrı gerçekleştireceğiz.
# iki katmanlı bir YSA olduğu için forward propagation'da A1, Z1, A2 ve Z2 değerlerini hesaplıyoruz.
# backward prop.'da yeniden hesaplama yapmamak ve hazır kullanmak için cache alıyoruz.
def forward_propagation(X, parameters):
    # initialize_parameters() fonksiyonundan gelen değerler.
    W1 = parameters['W1']
    b1 = parameters['b1']
    W2 = parameters['W2']
    b2 = parameters['b2']
    
    # (1) numaralı formül.
    Z1 = np.dot(W1, X) + b1
    
    # (2) numaralı formül.
    A1 = np.tanh(Z1)
    
    # (3) numaralı formül.
    Z2 = np.dot(W2, A1) + b2
    
    # (4) numaralı formül.
    A2 = sigmoid(Z2)
    
    cache = {"Z1": Z1,
             "A1": A1,
             "Z2": Z2,
             "A2": A2}
    
    return A2, cache

#====================
# cost değerinin hesaplanması.
def compute_cost(A2, Y):
    # veri miktarı.
    m = Y.shape[1] 
    
    # (5) ve (6) numaralı formüller.
    logprobs = np.multiply(np.log(A2), Y) + np.multiply((1 - Y), np.log(1 - A2))
    cost = - np.sum(logprobs) / m
    
    # vektörel sonucu skaler sonuca döndürmek için.
    # örnek: [[17]] --> 17 
    cost = float(np.squeeze(cost))  
    
    # sonuç
    return cost

#====================
# backward propagation
def backward_propagation(parameters, cache, X, Y):
    # veri miktarı.
    m = X.shape[1]
    
    # daha önce hesaplanan değerlerin cache'ten çekilmesi.
    W1 = parameters['W1']
    W2 = parameters['W2']    
    A1 = cache['A1']
    A2 = cache['A2']

    # türevlerin hesaplanması.
    # (7) numaralı formül.
    dZ2 = A2 - Y
    
    # (8) numaralı formül.
    dW2 = (1 / m) * np.dot(dZ2, A1.T)
    
    # (9) numaralı formül.
    db2 = (1 / m) * np.sum(dZ2, axis=1, keepdims=True)
    
    # (10) numaralı formül.
    dZ1 = np.multiply(np.dot(W2.T, dZ2), 1 - np.power(A1, 2))
    
    # (11) numaralı formül.
    dW1 = (1 / m) * np.dot(dZ1, X.T)
    
    # (12) numaralı formül.
    db1 = (1 / m) * np.sum(dZ1, axis=1, keepdims=True)
    
    # sonuç
    grads = {"dW1": dW1,
             "db1": db1,
             "dW2": dW2,
             "db2": db2}
    
    return grads

#====================
# gradient descent'ın uygulanması.
def update_parameters(parameters, grads, learning_rate):
    # daha önce hesaplanan değerlerin çekilmesi.
    W1 = parameters['W1']
    b1 = parameters['b1']
    W2 = parameters['W2']
    b2 = parameters['b2']
    dW1 = grads['dW1']
    db1 = grads['db1']
    dW2 = grads['dW2']
    db2 = grads['db2']

    # yeni w ve b değerlerinin üretilmesi.
    W1 = W1 - learning_rate * dW1
    b1 = b1 - learning_rate * db1
    W2 = W2 - learning_rate * dW2
    b2 = b2 - learning_rate * db2
    
    # sonuç.
    parameters = {"W1": W1,
                  "b1": b1,
                  "W2": W2,
                  "b2": b2}
    
    return parameters

#====================
# oluşturduğumuz tüm fonksiyonları bir model fonksiyonunda bir araya getirelim.
def nn_model(X_train, Y_train, X_test, Y_test, hidden = 4, epoch = 3000, learning_rate = 1.2, print_cost=False):
    costs = []
    
    print("X_train boyutu: ", X_train.shape)
    print("Y_train boyutu: ", Y_train.shape)
    print("X_test boyutu: ", X_test.shape)
    print("Y_test boyutu: ", Y_test.shape)
    
    # başlangıç parametreleri belirleniyor...
    parameters = initialize_parameters(X_train, Y_train, hidden)

    # girilen dönem (epoch) sayısı kadar hesaplama yapılacak.
    for i in range(0, epoch):
        
        # forward propagation.
        A2, cache = forward_propagation(X_train, parameters)
        
        # cost hesaplanıyor.
        cost = compute_cost(A2, Y_train)
 
        # backward propagation.
        grads = backward_propagation(parameters, cache, X_train, Y_train)
 
        # gradient descent
        # yeni w ve b değerlerinin üretilmesi.
        parameters = update_parameters(parameters, grads, learning_rate)
        
        # 100 dönemde bir cost kaydediliyor, böylece ne kadar azaldığını grafikten izleyebileceğiz.
        if i % 100 == 0:
            costs.append(cost)
            
        # 100 dönemde bir cost değerinin ekrana basılması.
        if print_cost and i % 100 == 0:
            print ("Epoch: %i, Cost: %f" %(i, cost))

    # train ve test veri setleri ile tüm tahminler hesaplanıyor.
    Y_hat_test = predict(parameters, X_test)
    Y_hat_train = predict(parameters, X_train)

    # yapılan tahminler ile gerçek veriler karşılaştırılıp, doğruluk oranı hesaplanıyor.
    print("Train Accuracy: {} %".format(100 - np.mean(np.abs(Y_hat_train - Y_train)) * 100))
    print("Test Accuracy: {} %".format(100 - np.mean(np.abs(Y_hat_test - Y_test)) * 100))
    
    # bulunan tüm sonuçlar bir sözlükte toplanıyor.
    d = {"costs": costs,
         "Y_hat_test": Y_hat_test, 
         "Y_hat_train" : Y_hat_train, 
         "parameters" : parameters, 
         "learning_rate" : learning_rate,
         "epoch": epoch}
    
    return d

#====================
# tahmin işlemi.
def predict(parameters, X):
    # tahmin edilecek y değerleri (y şapka).
    Y_hat = np.zeros((1,X.shape[1]))
    
    A2, cache = forward_propagation(X, parameters)
    pred = np.round(A2)
    
    for i in range(pred.shape[1]):
        if (A2[0,i] > 0.5): 
            Y_hat[0,i] = 1.
        else:
            Y_hat[0,i] = 0.
    
    return Y_hat
```

Şimdi kullanabileceğimiz fonksiyonlar hazır olduğuna göre, örnek bir veri seti için 2 katmanlı YSA modelimizi çalıştırabiliriz.  Amaç sadece 2 katmanlı YSA'nın nasıl çalıştığını ve yukarıda anlatılan formüllerin kullanımını göstermek olduğundan, veri analizi ve veri setinin yorumlanması vb. ayrıntılara girilmemiştir. 2 katmanlı YSA ile lojistik regresyon sonuçlarını karşılaştırmak için, daha önceki derste kullandığımız verisetlerinin aynısını kullanacağız.

## ÖRNEK 1

```python
churn_data = pd.read_csv("datasets/churn_data.csv")
df1 = churn_data.copy()

from sklearn.model_selection import train_test_split
X = df1.drop(['churn'], axis=1)
Y = df1['churn']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=30)

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

X_train_scaled = X_train_scaled.T
X_test_scaled = X_test_scaled.T

Y_train = Y_train.to_numpy().reshape(1,X_train_scaled.shape[1])
Y_test = Y_test.to_numpy().reshape(1,X_test_scaled.shape[1])

d1 = nn_model(X_train_scaled, Y_train, X_test_scaled, Y_test, 
              hidden = 11, epoch = 12000, learning_rate = 0.007, print_cost=True)

import matplotlib.pyplot as plt

plt.plot(d1['costs'])
plt.xlabel('Epoch')
plt.ylabel('Cost')
plt.show()
```

```python
# veri setinde olmayan resimlerle, modelin denenmesi.
from PIL import Image
from matplotlib.pyplot import imshow

for kedi in range(5):
    # resim dosyasının dizini
    fname = "images/kedi" + str(kedi) +".jpg" 

    # resim dosyası açılır, üstte belirlenen num_px değerine resize olur.
    img = Image.open(fname)
    im_resized = img.resize(size=(num_px, num_px))
    
    # resmin piksellerinin standardizasyonu
    arr_img = np.asarray(im_resized)
    arr_img = arr_img/255.
    
    # resmin piksellerinin predict() fonksiyonunda kullanılabilir hale getirilmesi.
    my_image = arr_img.reshape((1, num_px*num_px*3)).T
    
    # tahmin işlemi.
    my_predicted_image = predict(d3["parameters"], my_image)

    # tahmin etiketinin belirlenmesi.
    if np.squeeze(my_predicted_image):
        pred = "KEDİ OLDUĞUNU"
    else:
        pred = "KEDİ OLMADIĞINI"
    
    # sonuç.
    print("Modelimiz bu resmin", pred ,"söyledi")

    # tahmin edilen resmin ekranda gösterilmesi.
    plt.figure()
    plt.imshow(np.asarray(im_resized))
    plt.show()
```

Bu modelimiz de, lojistik regresyonda olduğu gibi 5 resmin 3'ünü doğru tahmin etti.
